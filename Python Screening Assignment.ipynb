Decorators
In decorators we basically say that inside a function we can also give parameter as a function instead of a specific variable or message kind of things.

```python
### Decorators
def main_class(func):
    
    def sub_class():
        print("Good Morning")
        func()
        print("Good Afternoon")
    return sub_class()
```

```python
@main_class
def final_class():
    print("Good Evening")
```

# Multiple Inheritance

```python
class A:
    def method1(self):
        print("A class method is called")
        
class B(A):
    def method1(self):
        print("B class method is called")
    def method2(self):
        print("B class method is called")

class C(A):
    def method1(self):
        print("C class method is called")
        
class D(B,C):
    def method1(self):
        print("D class method is called")
        C.method1(self)
        B.method1(self)
        A.method1(self)
```

```python
d = D()
```

```python
d.method1()
```

```python
#method2 of class B
d.method2()
```

```python
## call class B method 1
B.method1(d)
```

```python
## call class A method 1
A.method1(d)
```

# Abstraction

```python
from abc import ABC, abstractmethod
class car(ABC):
    def mileage(self):
        pass
    
class Creta(car):
    def mileage(self):
        print("The milege is 22 kmph")
        
class Bmw(car):
    def mileage(self):
        print("The milege is 15 kmph")
        
class Jeep(car):
    def mileage(self):
        print("The milege is 17 kmph")
        
class Duster(car):
    def mileage(self):
        print("The milege is 11 kmph")
```

```python
t= Creta()   
t.mileage()   
  
a = Bmw()   
a.mileage()   
  
a1 = Jeep()   
a1.mileage() 

f = Duster()   
f.mileage()  
```

A class is called an Abstract class if it contains one or more abstract methods. An abstract method is a method that is declared, but contains no implementation. Abstract classes may not be instantiated, and its abstract methods must be implemented by its subclasses
